```python
import pandas as pd
import numpy as np

import collections 
from collections import Counter 
import string 

import matplotlib.pyplot as plt
from wordcloud import WordCloud, ImageColorGenerator

import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk import download

from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, chi2
from sklearn import metrics
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier

print("Finished!")
```

<h1> Ingest + prep data </h1>

```python
# Read in data and create data frame; call it "reviews"
reviews = pd.read_csv("company_reviews.csv")

# Convert status into binary numerical format and clean up space around name
status_dict = {'Former Employee': 0,'Former Employee ': 0,'Current Employee': 1, 'Current Employee ':1} 
reviews['status_num'] = [status_dict[item] for item in reviews.status]
```

<h1> Term frequency: wordclouds and frequency tables  </h1>

```python
# What are the most common words in the corpus?
# Generate wordclouds and frequency tables
# Resource: https://www.datacamp.com/community/tutorials/wordcloud-python [uses another verion of same wine reviews dataset]

# Join all summaries into a single text blob
text = " ".join(summary for summary in reviews.summary)

# Generate a word cloud
wordcloud = WordCloud(background_color="white").generate(text)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()

# Helpful? Pros and cons of word clouds? Additional steps to clean the text?
```

```python
# Generate word frequency table and view it as a dataframe
counts = collections.Counter(text.split()) # Create counts
counts_df = pd.DataFrame.from_dict(counts, orient='index').reset_index() # Convert into dataframe
counts_df.columns = ['word', 'frequency']  # Rename columns
counts_df.sort_values(by=['frequency'], inplace=True, ascending = False) # Sort by frequency

counts_df # View the dataframe

# What do you notice?
# Are all words in our corpus of equal value?
# Do you want to drop some?  Do some other cleaning?
```

```python
# Do the following in the original dataframe, in a new "summary_cleaned" column: 
# Tokenize; lowercase all words; drop numbers, punctuation, symbols

# Tokenize
reviews['summary_cleaned_tok'] = reviews['summary'].apply(word_tokenize)
```

```python
# Lowercase
reviews['summary_cleaned_tok'] = reviews['summary_cleaned_tok'].apply(lambda x: [word.lower() for word in x])
```

```python
# Stopwords
download('stopwords')  
stop_words = stopwords.words('english')
#stop_words # Look at built-in list of stopwords; note that they're lowercase

# Optional: add words to built-in stopwords 
stop_words.append('i') # To add an individual word

# Or
# stop_to_drop = ['word1','word2'] # To add a list of words
# stop_words.extend(stop_to_drop)

#stop_words.remove('i')
stop_words
```

```python
# Drop stopwords
reviews['summary_cleaned_tok'] = reviews['summary_cleaned_tok'].apply(lambda x: [word for word in x if word not in stop_words])
```

```python
# Remove punctuation
punc = string.punctuation
reviews['summary_cleaned_tok'] = reviews['summary_cleaned_tok'].apply(lambda x: [word for word in x if word not in punc])
```

```python
# Remove numbers
# https://docs.python.org/3/library/string.html
digits = string.digits
reviews['summary_cleaned_tok'] = reviews['summary_cleaned_tok'].apply(lambda x: [word for word in x if word not in digits])
```

```python
# Re-string [to be in format required for wordcloud]
reviews['summary_cleaned_str'] = [' '.join(map(str, l)) for l in reviews['summary_cleaned_tok']]
reviews
```

```python
# Re-do the wordcloud with the cleaned text

# Join all descriptions into a single text blob
text_cleaned = " ".join(summary for summary in reviews.summary_cleaned_str)

# Generate a word cloud
wordcloud = WordCloud(background_color="white").generate(text_cleaned)

# Display the generated image:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```

```python
# Re-generate word frequency table and view it as a dataframe
counts = collections.Counter(text_cleaned.split()) # Create counts
counts_df = pd.DataFrame.from_dict(counts, orient='index').reset_index() # Convert into dataframe
counts_df.columns = ['word', 'frequency']  # Rename columns
counts_df.sort_values(by=['frequency'], inplace=True, ascending = False) # Sort by frequency

counts_df # View the dataframe -- do you want to go back upstream and do any more cleaning?
```

<h1> Prediction via logistic regression, decision tree, random forest </h1>

```python
# Preliminary step: look at distribution of former vs. current employees
reviews['status_num'].value_counts().plot(kind='bar')

# Class imbalance?  Could rebalance.
```

```python
# Convert each review into a bag-of-words (BoW) vector.
vectorizer_bow = CountVectorizer() # Create vectorizer tool
bow_features = vectorizer_bow.fit_transform(reviews['summary_cleaned_str']).toarray() # Run on text column
bow_features.shape # Look at shape (num rows, num columns)
```

```python
bow_features
```

```python
# Create training and test sets (80% training; 20% test)
X = reviews['summary_cleaned_str']
y = reviews['status_num']

train_X, test_X, train_y, test_y = train_test_split(X, y, test_size = 0.2, random_state=200)
```

```python
# Run your regression on your training set only
pipeline = Pipeline([('vect', vectorizer_bow),
                     ('chi',  SelectKBest(chi2, k='all')),
                     ('clf', LogisticRegression(random_state=0, max_iter=1000))]) 

model = pipeline.fit(train_X, train_y)

# Generate F-1 score
metrics.f1_score(test_y, model.predict(test_X))
```

```python
# Decision tree
pipeline = Pipeline([('vect', vectorizer_bow),
                     ('chi',  SelectKBest(chi2, k='all')),
                     ('clf', DecisionTreeClassifier(random_state=0))]) 

model = pipeline.fit(train_X, train_y)

# Generate F-1 score
metrics.f1_score(test_y, model.predict(test_X))
```

```python
# Random forest
pipeline = Pipeline([('vect', vectorizer_bow),
                     ('chi',  SelectKBest(chi2, k='all')),
                     ('clf', RandomForestClassifier(random_state=0))]) 

model = pipeline.fit(train_X, train_y)

# Generate F-1 score
metrics.f1_score(test_y, model.predict(test_X))
```
