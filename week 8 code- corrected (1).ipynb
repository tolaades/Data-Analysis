<h1> Ingest + prep data </h1>

```python
# Define a function for calculating mean squared prediction error 
def mean_sq_prediction_error(y_actual, y_predicted):
    mpe = ((y_actual - y_predicted)**2).mean()
    print("Mean prediction error is % s" % round(mpe, 5)) 
```

```python
# Read in data and create data frame; call it "reviews"
reviews = pd.read_csv("company_reviews.csv")
```

```python
# Convert status into binary numerical format and clean up space around name
status_dict = {'Former Employee': 0, 'Current Employee': 1} 
reviews['status_num'] = [status_dict[item.strip()] for item in reviews.status]
```

<h1> Continuous dependent variable </h1>

```python
# Isolate your X's and y -- continuous
independent = ['month', 'year', 'state', 'status_num', 'likes']
dependent = 'score'
```

```python
# Split your data frame into training and test sets: 80% training; 20% test
X = pd.get_dummies(reviews[independent], drop_first=True)
y = reviews[dependent]
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size = 0.2, random_state=200)
```

<h2> -- Decision tree </h2>

```python
# Build the tree
tree_cont = DecisionTreeRegressor()
tree_cont = tree_cont.fit(train_X, train_y)
```

```python
# Show the tree
text_representation = tree.export_text(tree_cont)
print(text_representation)
```

```python
# Assign predicted values to an object
y_tree_predicted = tree_cont.predict(test_X)
```

```python
# Convert comparison into a data frame and eyeball it
predictions = pd.DataFrame({'score': test_y,
                    'predicted_score': y_tree_predicted})

predictions
```

```python
# Compute the mean squared prediction error-- how well did the model perform?
mean_sq_prediction_error(predictions['score'], predictions['predicted_score'])
```

```python
# You can compare this performance directly to your best logistic regression performance by mean squared prediction error.
```

<h2> -- Random forest </h2>

```python
# Build the forest
rf_cont = RandomForestRegressor(max_depth=2, random_state=0)
rf_cont = rf_cont.fit(train_X, train_y)

# Assign predicted values to an object
y_rf_cont_predicted = rf_cont.predict(test_X)
    
# Convert comparison into a data frame
predictions = pd.DataFrame({ 'score' : test_y,
                    'predicted_score':y_rf_cont_predicted})

# Compute the mean squared prediction error
mean_sq_prediction_error(predictions['score'], predictions['predicted_score'])
```
